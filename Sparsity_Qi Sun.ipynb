**1. Jokester**

The dataset that I picked is jester-data-3: (2.1MB) Data from 24,938 users who have rated between 15 and 35 jokes, a matrix with dimensions 24,938 X 101.

* Ratings are real values ranging from -10.00 to +10.00 (the value "99" corresponds to "null" = "not rated").

* One row per user

* The first column gives the number of jokes rated by that user. The next 100 columns give the ratings for jokes 01 - 100.


```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/susanqisun/DAV6300/main/jester-data-3.csv', header=None)
df
```

```python
# check the shape for the dataframe
print(df.shape)
```

```python
#check missing value
df.isnull().sum()
```

### Data Preprocessing

Dataset contains no column names. The first column gives the number of jokes rated by that user. The next 100 columns give the ratings for jokes 01 - 100. 


#### 1. Rename column

```python
#0th column would be renamed to freq
df.rename(columns = {0: 'freq'}, inplace = True)
df.head()
```

#### 2. Replace all 99.0 ratings as 0

The value "99" corresponds to "null" = "not rated", which means jokes were not rated by users. I'll replace them as 0.

```python
df02 = df.replace(99.0, 0)
df02.head()
```

#### 3. Transform from wide to long

```python
df03 = df02.copy()
df03.index = np.arange(1, len(df02) + 1)
df03
```

```python
# Save the 'freq' coloumn that I don't want to scale
freq = df03['freq'].reset_index()

freq.rename(columns = {'index': 'UserId'}, inplace = True)

freq
```

```python
#Extract ratings columns
ratings = df03.iloc[:, 1:]
ratings
```

```python
df04 = ratings.stack().reset_index()
df04.columns=['UserId', 'jokeID', 'rating']

df04
```

```python
# add freq column
# merge together
df05 = pd.merge(left=freq, right=df04, how='outer')
df05
```

#### Sparsity of a matrix:

Sparsity of Matrix = Number of Empty cells / Total Number of cells

```python
(df05['rating']==0).sum() / 2493800
```

It means that 75.34% of cells are empty. 

#### 4. Normalize ratings

In this dataset, there are positive and negative ratings. For example, for jokes 01, the lowest rating is -9.66, and the highest rating is 8.88. I'll normalize ratings columns by using StandardScalar, so it's distribution will have mean of 0 and standard deviation of 1. 

```python
ratings.describe()
```

```python
#Fit StandardScaler into ratings
from sklearn.preprocessing import StandardScaler

df05[['rating']] = StandardScaler().fit_transform(df05[['rating']])
df05
```
